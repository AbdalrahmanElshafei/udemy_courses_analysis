### Number of subscribers per year

```python
df.groupby('year')['num_subscribers'].sum()
```

```python
df.groupby('year')['num_subscribers'].sum().plot(kind='barh', color='skyblue')

plt.title('Total Number of Subscribers per Year')   
plt.xlabel('Number of Subscribers')                                
plt.ylabel('Year')                 
plt.tight_layout()
plt.show()
```
